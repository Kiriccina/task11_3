```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import root_mean_squared_error
import pandas as pd
import numpy as np

def get(name_file):

  file = pd.read_csv(name_file)

  features = file[['Year', 'Engine HP', 'Engine Cylinders', 'highway MPG', 'city mpg', 'Popularity']]
  target = file['MSRP']
  features = features.fillna(features.mean())
  features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.25, random_state=42)

  model = LinearRegression()
  model.fit(features_train, target_train)
  r = model.predict(features_test)

  rmse = root_mean_squared_error(target_test, r)

  stupid = target_train.mean()
  dumb_model = np.full(len(target_test), target_train.mean())
  drmse = root_mean_squared_error(target_test, dumb_model)

  return (rmse, drmse, drmse - rmse)
```
